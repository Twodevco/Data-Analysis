## jobs data analysis (2020 & 2021)

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data=pd.read_csv("wuzzaf_2021.csv")
data.sample(10)
```

### 3 Description column will divided it into 2 differ columns

```python
job_type = []
job_level = []
description=data["job_details"]
for i in description:
    if "Full Time" in str(i):
        job_type.append("Full Time")
    elif "Part Time" in str(i):
        job_type.append("Part Time")
    elif "Internship" in str(i):
        job_type.append("Internship")
    elif "Contract" in str(i):
        job_type.append("Contract")
    elif "Shift Based" in str(i):
        job_type.append("Shift Based")
    elif "Work From Home" in str(i):
        job_type.append("Work From Home")
    else:
        job_type.append(np.nan)
        
    if "Experienced" in str(i):
        job_level.append("Experienced")
    elif "Entry Level" in str(i):
        job_level.append("Entry Level")
    elif "Manager" in str(i):
        job_level.append("Manager")
    elif "Student" in str(i):
        job_level.append("Student")
    elif "Sinior Management" in str(i):
        job_level.append("Sinior Management")
    else :
        job_level.append(np.nan)
data["type"] = job_type
data["level"] = job_level
```

```python
data_2021=data
data_2021.head()
```

```python
data_2020 = pd.read_csv("wuzzaf_2020.csv")
data_2020
```

###  Cleaning

```python
data_2020_copy = data_2020.copy()
```

```python
data_2020_copy
```

### * Drop (#page URL and job_link)
### * Fixed some problems with the column (location)
### * Rename Columns (title , company-name , job-link_link , short-description- )
 

```python
data_2020_copy = data_2020_copy.drop(["#PAGE_URL","job-link","job-link_link"] , axis=1)
for i in range(len(data_2020_copy["location"])):
    data_2020_copy["location"][i] = data_2020_copy["location"][i].replace(",","")
data_2020_copy.rename(columns={"title":"job_title" , "company-name":"company_name" , "short-description-":"job_details"} , inplace=True)
```

```python
job_type = []
job_level = []
description = data_2020_copy["job_details"]
for i in description:
    if "Full Time" in str(i):
        job_type.append("Full Time")
    elif "Part Time" in str(i):
        job_type.append("Part Time")
    elif "Internship" in str(i):
        job_type.append("Internship")
    elif "Contract" in str(i):
        job_type.append("Contract")
    elif "Shift Based" in str(i):
        job_type.append("Shift Based")
    elif "Work From Home" in str(i):
        job_type.append("Work From Home")
    else:
        job_type.append(np.nan)
        
    if "Experienced" in str(i):
        job_level.append("Experienced")
    elif "Entry Level" in str(i):
        job_level.append("Entry Level")
    elif "Manager" in str(i):
        job_level.append("Manager")
    elif "Student" in str(i):
        job_level.append("Student")
    elif "Sinior Management" in str(i):
        job_level.append("Sinior Management")
    else :
        job_level.append(np.nan)
    
data_2020_copy["type"] = job_type
data_2020_copy["level"] = job_level
```

```python
data = data_2021.append(data_2020_copy).dropna()
```

```python
data=data.drop(['location'],axis=1)
```

```python
Group_list=[]
for x in data.job_title:
    if("IT" in x):
        Group_list.append('IT Category')
    elif('Software'or "Full Stack"or"Developer"or"End"or"end" in x):
        Group_list.append('Software Engneering ')
    else:
        Group_list.append(data.job_title)  
```

```python
data['Standard_Job_Title']=Group_list
```
